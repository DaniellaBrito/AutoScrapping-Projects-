```python
from autoscraper import AutoScraper
```

Scraping to find the price and details of headphones .

```python
url ='https://www.amazon.in/s?k=headphones%E0%A4%82&crid=1GRLMC80PQIK3&sprefix=head%2Caps%2C308&ref=nb_sb_ss_ts-doa-p_1_4'
```

```python
wanted_list =['₹999','pTron Bassbuds Plus True Wireless Bluetooth 5.0 Headphones with Deep Bass, Made in India, IPX4 Water/Sweat Resistant, Passive Noise Canceling Earbuds, Digital Display Case & Built-in HD Mic - (Black)','4,931']
```

```python
scraper = AutoScraper()
result = scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_ejdo':'Title','rule_kfrh':'stars'})
scraper.keep_rules({'rule_ejdo':'Title','rule_kfrh':'stars'})
scraper.save('amazon-search')
```

Scraping to find the price and details of iphones in market.

```python
amazon_url="https://www.amazon.in/s?k=iphones&ref=nb_sb_noss_2"

wanted_list=["₹50,999","New Apple iPhone 11 (64GB) - Black"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_fnit':'Title','rule_q098':'Price'})
scraper.keep_rules(['rule_fnit','rule_q098'])
scraper.save('amazon-search')
```

```python
results=scraper.get_result_similar('https://www.amazon.in/s?k=iphones&ref=nb_sb_noss_2',group_by_alias=True)

results['Price']
```
